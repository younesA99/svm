```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Données
fcj = np.array([8.06, 14.59, 28.24, 39.36, 44.21, 44.7])
age = np.array([3, 7, 28, 90, 180, 360])

# Redimensionner les données
X = age.reshape(-1, 1)
y = fcj

# Ajuster un modèle de régression linéaire
model = LinearRegression()
model.fit(X, y)

# Prédiction de la résistance à différents âges
age_pred = np.array([1, 30, 60, 120, 240, 360]).reshape(-1, 1)
fcj_pred = model.predict(age_pred)

# Visualisation des résultats
plt.scatter(age, fcj, label='Données réelles')
plt.plot(age_pred, fcj_pred, label='Régression linéaire')
plt.xlabel('Âge (jours)')
plt.ylabel('Fcj')
plt.legend()
plt.show()

# Mesure de la performance du modèle
mse = mean_squared_error(y, model.predict(X))
r2 = r2_score(y, model.predict(X))

print(f'Mean Squared Error (MSE): {mse}')
print(f'R-squared (R²): {r2}')
```

```python
import pandas as pd

# Création d'un DataFrame avec les colonnes ID, Y_test et predictions
results_df = pd.DataFrame({'y': y, 'fcj_pred': fcj_pred})

# Affichage des cinq premières lignes du DataFrame
print(results_df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Création du nuage de points

sns.scatterplot( x=y, y=fcj_pred, alpha=0.6)

# Tracé de la droite y = x pour comparer
plt.plot(results_df['y'], results_df['y'], color='red', label='y=x')

plt.xlabel('Valeurs réelles', fontsize=14)
plt.ylabel('Prédictions', fontsize=14)
plt.title('Valeurs réelles vs Prédictions (ensemble de test)', fontsize=17)
plt.legend()
plt.show()
```

```python
# Prédiction de la résistance à 14 jours
jours_pred_14 = np.array([14]).reshape(-1, 1)
resistance_pred_14 = model.predict(jours_pred_14)

# Prédiction de la résistance à 28 jours
jours_pred_28 = np.array([28]).reshape(-1, 1)
resistance_pred_28 = model.predict(jours_pred_28)

# Afficher les prédictions
print(f"Prédiction de la résistance à {jours_pred_14[0][0]} jours: {resistance_pred_14[0]} MPa")
print(f"Prédiction de la résistance à {jours_pred_28[0][0]} jours: {resistance_pred_28[0]} MPa")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score

# Données
fcj = np.array([8.06, 14.59, 28.24, 39.36, 44.21, 44.7])
age = np.array([3, 7, 28, 90, 180, 360])

# Redimensionner les données
X = age.reshape(-1, 1)
y = fcj

# Appliquer une régression polynomiale (d'ordre 2 dans cet exemple)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, y)

# Prédiction de la résistance à différents âges
age_pred = np.array([1, 30, 60, 120, 240, 360]).reshape(-1, 1)
X_pred = poly.transform(age_pred)
fcj_pred = model.predict(X_pred)

# Visualisation des résultats
plt.scatter(age, fcj, label='Données réelles')
plt.plot(age_pred, fcj_pred, label='Régression polynomiale (ordre 2)')
plt.xlabel('Âge (jours)')
plt.ylabel('Fcj')
plt.legend()
plt.show()

# Mesure de la performance du modèle
mse = mean_squared_error(y, model.predict(X_poly))
r2 = r2_score(y, model.predict(X_poly))

print(f'Mean Squared Error (MSE): {mse}')
print(f'R-squared (R²): {r2}')
```

```python
import pandas as pd

# Création d'un DataFrame avec les colonnes ID, Y_test et predictions
results_df = pd.DataFrame({'y': y, 'fcj_pred': fcj_pred})

# Affichage des cinq premières lignes du DataFrame
print(results_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Création du nuage de points

sns.scatterplot( x=y, y=fcj_pred, alpha=0.6)

# Tracé de la droite y = x pour comparer
plt.plot(results_df['y'], results_df['y'], color='red', label='y=x')

plt.xlabel('Valeurs réelles', fontsize=14)
plt.ylabel('Prédictions', fontsize=14)
plt.title('Valeurs réelles vs Prédictions (ensemble de test)', fontsize=17)
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger le fichier Excel
file_path = 'Concrete_local data.xlsx'  # Remplacez par le chemin de votre fichier
data = pd.read_excel(file_path)

# Sélectionner et renommer les colonnes comme précédemment...
data_local = data[['Temperature ambiante', 'Ciment CPJ45 (kg/m3)', 'Gravier 8/12 (kg/m3)', 'Gravier 4/8 (kg/m3)',
                      'Sable 0/4 (kg/m3)', 'E/C', 'G/S', 'Affaissement (cm)', 'Durée de durcissement (7jrs, 14jrs ou 28jrs)',
                      'Masse  eprouvette (Kg)', 'Résistance meca (MPa)']]

# Combiner les colonnes 'Gravier 8/12 (kg/m3)' et 'Gravier 4/8 (kg/m3)'
data_local['Gravier (kg/m3)'] = data_local['Gravier 8/12 (kg/m3)'] + data_local['Gravier 4/8 (kg/m3)']

# Supprimer les colonnes 'Gravier 8/12 (kg/m3)' et 'Gravier 4/8 (kg/m3)'
data_local = data_local.drop(['Gravier 8/12 (kg/m3)', 'Gravier 4/8 (kg/m3)'], axis=1)

# Renommer les colonnes
data_local.columns = ['Temperature', 'Cement', 'Sand', 'Water/Cement Ratio', 'Gravel/Sand Ratio',
                         'Slump (cm)', 'Curing Time', 'Specimen Mass (Kg)', 'Compressive Strength (MPa)', 'Gravel (kg/m3)']

# Réorganiser les colonnes avec "Compressive Strength (MPa)" en dernière position
columns_order = ['Temperature', 'Cement', 'Gravel (kg/m3)', 'Sand', 'Water/Cement Ratio', 'Gravel/Sand Ratio',
                 'Slump (cm)', 'Curing Time', 'Specimen Mass (Kg)', 'Compressive Strength (MPa)']
data_filtered = data_local[columns_order]

# Tracer le nuage de points et les courbes de variation de la résistance en fonction des autres variables
plt.figure(figsize=(16, 24))
for i, column in enumerate(data_filtered.columns[:-1]):
    plt.subplot(4, 2, i+1)
    sns.scatterplot(data=data_filtered, x=column, y='Compressive Strength (MPa)')
    sns.lineplot(data=data_filtered, x=column, y='Compressive Strength (MPa)', ci=None, color='red')
    plt.title(f'Compressive Strength vs {column}')
    plt.xlabel(column)
    plt.ylabel('Compressive Strength (MPa)')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Définir les variables indépendantes (features) et dépendantes (target)
x = data_filtered.drop('Compressive Strength (MPa)', axis=1)
y = data_filtered['Compressive Strength (MPa)']

# Diviser les données en ensembles de formation et de test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Normalisation des données
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

# Définir les hyperparamètres pour Random Forest
param_dist = {
    'n_estimators': np.arange(1, 21, 1),
    'max_depth': [None] + list(np.arange(1, 21, 1))
}

# Créer un objet de régression RandomForest
rf = RandomForestRegressor(random_state=42)

# Recherche aléatoire avec validation croisée
random_search = RandomizedSearchCV(estimator=rf, param_distributions=param_dist,
                                   n_iter=20, cv=5, scoring='neg_mean_squared_error',
                                   verbose=1, n_jobs=-1, random_state=42)

# Adapter le modèle aux données d'entraînement normalisées
random_search.fit(x_train_scaled, y_train)

# Afficher les meilleurs paramètres
print("Meilleurs paramètres:", random_search.best_params_)

# Prédire sur l'ensemble de test
y_pred = random_search.best_estimator_.predict(x_test_scaled)

# Calculer l'erreur quadratique moyenne sur l'ensemble de test
mse = mean_squared_error(y_test, y_pred)
print("Erreur quadratique moyenne:", mse)

# Entraîner le modèle avec les meilleurs paramètres
best_params = random_search.best_params_
rf_model = RandomForestRegressor(**best_params, random_state=42)
rf_model.fit(x_train_scaled, y_train)

# Faire des prédictions sur l'ensemble de test
y_pred_test_rf = rf_model.predict(x_test_scaled)

# Calculer les métriques de performance pour l'ensemble de test
mae_test_rf = mean_absolute_error(y_test, y_pred_test_rf)
rmse_test_rf = mean_squared_error(y_test, y_pred_test_rf, squared=False)
r2_test_rf = r2_score(y_test, y_pred_test_rf)

# Faire des prédictions sur l'ensemble d'entraînement
y_pred_train_rf = rf_model.predict(x_train_scaled)

# Calculer les métriques de performance pour l'ensemble d'entraînement
mae_train_rf = mean_absolute_error(y_train, y_pred_train_rf)
rmse_train_rf = mean_squared_error(y_train, y_pred_train_rf, squared=False)
r2_train_rf = r2_score(y_train, y_pred_train_rf)

print("\nPerformance sur l'ensemble de test (Random Forest):")
print("Mean Absolute Error (MAE):", mae_test_rf)
print("Root Mean Squared Error (RMSE):", rmse_test_rf)
print("R-squared (R²):", r2_test_rf)

print("\nPerformance sur l'ensemble d'entraînement (Random Forest):")
print("Mean Absolute Error (MAE):", mae_train_rf)
print("Root Mean Squared Error (RMSE):", rmse_train_rf)
print("R-squared (R²):", r2_train_rf)

# Plot pour l'ensemble d'entraînement et de test
plt.figure(figsize=(10, 5))

# Entraînement
plt.scatter(y_train, y_pred_train_rf, color='blue', label='Train')
plt.plot([min(y_train), max(y_train)], [min(y_train), max(y_train)], color='red', linestyle='--', lw=2)

# Test
plt.scatter(y_test, y_pred_test_rf, color='green', label='Test')

plt.xlabel('True Values', fontdict={'fontsize': 12, 'fontweight': 'bold'})
plt.ylabel('Predicted Values', fontdict={'fontsize': 12, 'fontweight': 'bold'})
plt.title("Random Forest Model Performance - Training and Test Set", fontdict={'fontsize': 14, 'fontweight': 'bold'})
plt.legend()
plt.grid(True)

# Ajouter les valeurs de R² dans le graphique
plt.text(0.82, 0.17, f'R² (Train): {r2_train_rf:.3f}', fontsize=12, color='blue', transform=plt.gca().transAxes)
plt.text(0.82, 0.07, f'R² (Test) : {r2_test_rf:.3f}', fontsize=12, color='green', transform=plt.gca().transAxes)

plt.show()
```

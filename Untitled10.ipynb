```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVR
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

# Charger les donnÃ©es
TrainPd = pd.read_excel("Concrete_Data.xls")
TrainPd.columns = ['cement','blastFurnace','flyAsh','water','superplasticizer','courseAggregate','fineaggregate','age','strength']

# SÃ©parer les caractÃ©ristiques (X) de la cible (Y)
X = TrainPd.drop(["strength"], axis=1)
Y = TrainPd["strength"]

# Division des donnÃ©es en ensembles d'entraÃ®nement et de test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, random_state=1)

# CrÃ©ation du modÃ¨le SVR sans normalisation
model_unscaled = SVR(kernel='rbf')

# EntraÃ®nement du modÃ¨le sans normalisation
model_unscaled.fit(X_train, Y_train)

# CrÃ©ation du modÃ¨le SVR avec normalisation
scaler = MinMaxScaler()
X_train_normalized = scaler.fit_transform(X_train)
X_test_normalized = scaler.transform(X_test)

model_scaled = SVR(kernel='rbf')

# EntraÃ®nement du modÃ¨le avec normalisation
model_scaled.fit(X_train_normalized, Y_train)

# PrÃ©dictions sur l'ensemble de test sans normalisation
predictions_unscaled = model_unscaled.predict(X_test)

# PrÃ©dictions sur l'ensemble de test avec normalisation
predictions_scaled = model_scaled.predict(X_test_normalized)

# Visualisation des nuages de points avec la ligne parfaite
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(Y_test, predictions_unscaled)
plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], 'r--', lw=2)  # Ligne parfaite
plt.title('Sans Normalisation')
plt.xlabel('True Values')
plt.ylabel('Predictions')

plt.subplot(1, 2, 2)
plt.scatter(Y_test, predictions_scaled)
plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], 'r--', lw=2)  # Ligne parfaite
plt.title('Avec Normalisation')
plt.xlabel('True Values')
plt.ylabel('Predictions')

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import r2_score

# Listes pour stocker les performances
r2_scores_unscaled = []
r2_scores_scaled = []
epsilon_values = np.linspace(0.1, 100, 1000)
#epsilon_values = [0.1, 1, 10,100]  # Valeurs d'epsilon Ã  tester

# ModÃ¨le sans normalisation
for epsilon in epsilon_values:
    model_unscaled = SVR(kernel='rbf', epsilon=epsilon)
    model_unscaled.fit(X_train, Y_train)
    predictions_unscaled = model_unscaled.predict(X_test)
    r2_unscaled = r2_score(Y_test, predictions_unscaled)
    r2_scores_unscaled.append(r2_unscaled)

# ModÃ¨le avec normalisation
for epsilon in epsilon_values:
    model_scaled = SVR(kernel='rbf', epsilon=epsilon)
    model_scaled.fit(X_train_normalized, Y_train)
    predictions_scaled = model_scaled.predict(X_test_normalized)
    r2_scaled = r2_score(Y_test, predictions_scaled)
    r2_scores_scaled.append(r2_scaled)

# Affichage des performances en fonction de l'epsilon
plt.figure(figsize=(8, 5))

plt.plot(epsilon_values, r2_scores_unscaled, label='Sans Normalisation', marker='o')
plt.plot(epsilon_values, r2_scores_scaled, label='Avec Normalisation', marker='o')

plt.xlabel('Valeurs de Epsilon')
plt.ylabel('RÂ²')
plt.title("Performance du modÃ¨le SVR en fonction de l'epsilon\n Kernel=rbf ; C=1 ; epsilon_values = [0.1 Ã  100]")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# DÃ©tection des outliers avec la mÃ©thode de la plage interquartile (IQR)
outlier_threshold = 2.5  # Choisissez un seuil appropriÃ© pour identifier les outliers

for feature in X.columns:
    Q1 = X[feature].quantile(0.25)
    Q3 = X[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - outlier_threshold * IQR
    upper_bound = Q3 + outlier_threshold * IQR

    outliers = X[(X[feature] < lower_bound) | (X[feature] > upper_bound)]
    print(f"Outliers for {feature}: {len(outliers)}")
```

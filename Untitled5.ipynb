```python
import numpy as np
import pandas as pd
TrainPd=pd.read_excel("Concrete_Data.xls")
#renaming all featurses
TrainPd.columns = ['cement','blastFurnace','flyAsh','water','superplasticizer','courseAggregate','fineaggregate','age','strength']

#Verification des valeur manquantes
TrainPd.isna().sum()

# Split the data into features and target
X=TrainPd.drop(["strength"],axis=1)
Y=TrainPd["strength"]
Names=list(X.columns.values)

# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, random_state = 1)

#Normalisation des données
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
# Fit on training set only.
scaler.fit(X_train)
MinMaxScaler(copy=(), feature_range=(0,1))
# Apply transform to both the training set and the test set.
X_train_transformed = scaler.transform(X_train)
X_test_transoformed = scaler.transform(X_test)
```

```python
import numpy as np
from sklearn.svm import SVR
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline

# Liste des valeurs de C à tester
C_values = np.linspace(0.1,1000, 500)  # Créer 100 valeurs de 0.1 à 100

# Fixer les autres hyperparamètres
kernel = 'rbf'
gamma = 'scale'
epsilon = 0.1

# Stockage des performances pour chaque valeur de C
performance_scores = []

for C in C_values:
    # Initialisation du modèle avec la valeur de C actuelle
    svm_model = SVR(kernel=kernel, C=C, gamma=gamma, epsilon=epsilon)

    # Entraînement du modèle
    svm_model.fit(X_train_transformed, Y_train)

    # Prédiction sur les données de test
    predictions = svm_model.predict(X_test_transoformed)

    # Calcul de la performance (R² dans cet exemple)
    r2 = r2_score(Y_test, predictions)

    # Ajout du score de performance à la liste
    performance_scores.append(r2)

# Interpolation des données pour obtenir une courbe lisse
C_smooth = np.linspace(C_values.min(), C_values.max(), 300)  # Interpolation pour 300 valeurs
performance_smooth = make_interp_spline(C_values, performance_scores)(C_smooth)

# Affichage de la performance lissée en fonction de C
plt.plot(C_smooth, performance_smooth)
plt.xlabel('Valeurs de C')
plt.ylabel('Performance (R²)')
plt.title('Performance du modèle SVM en fonction de C (courbe lissée)')
plt.grid(True)
plt.show()
# Trouver les valeurs de C correspondant aux performances max, min, et aux percentiles
C_max = C_values[np.argmax(performance_scores)]
C_min = C_values[np.argmin(performance_scores)]
C_percentiles = np.percentile(C_values, [25, 50, 75], interpolation='nearest', axis=0)

# Récupérer les R² correspondant à ces valeurs de C
R2_max = np.max(performance_scores)
R2_min = np.min(performance_scores)
R2_percentiles = np.percentile(performance_scores, [25, 50, 75], interpolation='nearest', axis=0)

# Afficher les points sur la courbe
points_to_display = [(C_max, R2_max), (C_min, R2_min)]
percentile_points = [(C_percentiles[i], R2_percentiles[i]) for i in range(len(C_percentiles))]

for point in points_to_display:
    plt.scatter(point[0], point[1], color='red')
    plt.text(point[0], point[1], f'({point[0]:.2f}, {point[1]:.2f})', ha='right')

for percentile_point in percentile_points:
    plt.scatter(percentile_point[0], percentile_point[1], color='green')
    plt.text(percentile_point[0], percentile_point[1], f'({percentile_point[0]:.2f}, {percentile_point[1]:.2f})', ha='right')

plt.show()
```

```python
from sklearn.svm import SVR
from sklearn.metrics import r2_score
def train_evaluate_SVR(C_val, kernel_type, epsilon_val):
    svr = SVR(C=C_val, kernel=kernel_type, epsilon=epsilon_val)
    svr.fit(X_train_transformed, Y_train)
    predictions = svr.predict(X_test_transoformed)
    r2 = r2_score(Y_test, predictions)
    return r2
#C_values = [0.1, 1, 10, 100]
C_values = [100]
#kernel_types = ['linear', 'rbf']
kernel_types = ['rbf']
#epsilon_values = [0.1, 0.01, 0.001]
epsilon_values = [0.1, 0.01, 0.001]
results = {}

for C_val in C_values:
    for kernel_type in kernel_types:
        for epsilon_val in epsilon_values:
            key = f"C={C_val}, Kernel={kernel_type}, Epsilon={epsilon_val}"
            r2 = train_evaluate_SVR(C_val, kernel_type, epsilon_val)
            results[key] = r2

# Afficher les performances R² pour chaque combinaison d'hyperparamètres
for key, value in results.items():
    print(f"{key}: R² = {value}")
```

```python
import pandas as pd

# Définir les listes d'hyperparamètres
C_values = [0.1, 1, 10, 100]
kernel_types = ['linear', 'rbf']
epsilon_values = [0.1, 0.01, 0.001]

results = []

for C_val in C_values:
    for kernel_type in kernel_types:
        for epsilon_val in epsilon_values:
            key = f"C={C_val}, Kernel={kernel_type}, Epsilon={epsilon_val}"
            r2 = train_evaluate_SVR(C_val, kernel_type, epsilon_val)
            results.append([C_val, kernel_type, epsilon_val, r2])

# Créer un DataFrame Pandas à partir des résultats
columns = ['C', 'Kernel', 'Epsilon', 'R2']
results_df = pd.DataFrame(results, columns=columns)

# Afficher les résultats sous forme de matrice
print(results_df.pivot_table(index=['C', 'Kernel'], columns='Epsilon', values='R2'))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pivoter les données pour les préparer à l'affichage sous forme de heatmap
pivot_df = results_df.pivot_table(index=['Kernel', 'C'], columns='Epsilon', values='R2')

# Créer un diagramme de chaleur (heatmap)
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_df, annot=True, cmap='viridis', fmt=".3f")
plt.title('Performance (R²) de SVR avec différents hyperparamètres')
plt.xlabel('Valeurs de Epsilon')
plt.ylabel('Type de Kernel - Valeurs de C')
plt.show()
```

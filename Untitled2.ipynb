```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error

# Charger les données depuis le fichier Excel
data = pd.read_excel('Concrete_Data.xls')

# renaming all features
data.columns = ['cement', 'blastFurnace', 'flyAsh', 'water', 'superplasticizer', 'courseAggregate', 'fineaggregate', 'age', 'strength']

# Diviser les données en variables d'entrée (X) et variable cible (y)
X = data[['cement', 'blastFurnace', 'flyAsh', 'water', 'superplasticizer', 'courseAggregate', 'fineaggregate', 'age']]
y = data['strength']

# Normaliser les variables d'entrée
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Diviser les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'hidden_layer_sizes': [ (5,), (8,), (10,), (15,), (20,), (5,3), (3,5), (8,3), (15,3), (20,3), (30,30)],
    'activation': ['relu', 'tanh', 'identity', 'logistic'],
    'solver': ['adam', 'lbfgs', 'sgd'],
    'alpha': [0.01, 0.02, 0.05, 0.1, 0.2, 0.5],
    'learning_rate': ['constant', 'invscaling', 'adaptive'],
    'max_iter': [500,1000]
}


mlp = MLPRegressor(random_state=42)

gridSearch = GridSearchCV(mlp, param_grid,
                          cv=5, n_jobs=-1,
                          scoring='neg_mean_squared_error',
                          return_train_score=True)
gridSearch.fit(X_train, y_train.ravel())

print('Best Score (MSE)', -gridSearch.best_score_,
      '\nBest parameters: ', gridSearch.best_params_)
```

```python
# Listes pour stocker les MSE et les différentes configurations de hidden_layer_sizes
mse_values = []
hidden_layer_sizes_values = []

# Liste des différentes tailles de couches cachées à tester
hidden_layer_sizes_list = [(5,), (10,), (15,), (20,), (25,), (5, 5), (10, 10), (15, 15), (20, 20), (25, 25), (30, 25), (30, 30)]  # Ajoutez d'autres tailles si nécessaire

# Boucle pour entraîner le modèle avec différentes tailles de couches cachées
for size in hidden_layer_sizes_list:
    # Créer le modèle MLPRegressor avec la taille de couche cachée actuelleactivation='relu', solver='adam', alpha=0.01, learning_rate='constant', max_iter=500, random_state=42
    mlp = MLPRegressor(hidden_layer_sizes=size,
                      learning_rate='constant',
                      alpha=0.01,
                      activation='relu',
                      solver='adam',
                      random_state=42,
                      max_iter=1000)



    # Entraîner le modèle
    mlp.fit(X_train, y_train)

    # Faire des prédictions sur l'ensemble de test
    y_pred = mlp.predict(X_test)

    # Calculer le MSE et l'ajouter à la liste des MSE
    mse = mean_squared_error(y_test, y_pred)
    mse_values.append(mse)

    # Enregistrer la taille de la couche cachée correspondant à ce MSE
    if len(size) == 1:
        hidden_layer_sizes_values.append(str(size[0]) + '-Layer')
    else:
        hidden_layer_sizes_values.append(str(size[0]) + '-' + str(size[1]) + ' Layers')


# Visualiser la variation du MSE en fonction des différentes tailles de couches cachées
plt.figure(figsize=(10, 6))
bars = plt.bar(hidden_layer_sizes_values, mse_values, color='skyblue')
plt.xlabel('Configurations de Taille de Couche Cachée')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Variation du MSE en fonction des tailles de couches cachées')
plt.xticks(rotation=45, ha='right')  # Rotation des étiquettes pour une meilleure lisibilité
plt.grid(axis='y')

# Ajout des valeurs de MSE au-dessus de chaque colonne
for bar, mse in zip(bars, mse_values):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{mse:.4f}',
             ha='center', va='bottom', fontsize=8, color='black')

plt.tight_layout()
plt.show()

# Créer un DataFrame à partir des valeurs collectées
#results_df = pd.DataFrame({'Hidden_Layer_Configuration': hidden_layer_sizes_values, 'Mean_Squared_Error': mse_values})

# Afficher le DataFrame
#print(results_df)
```

```python
from sklearn.metrics import make_scorer

def plot_learning_curve_MSE(estimator, title, X, y, ylim=None, cv=None, n_jobs=None, train_sizes=np.linspace(.1, 1.0, 5)):
    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("MSE")

    # Définir une fonction de score personnalisée pour le MSE
    mse_scorer = make_scorer(mean_squared_error, greater_is_better=False)

    train_sizes, train_scores, test_scores = learning_curve(estimator, X, y, cv=cv, n_jobs=n_jobs,
                                                            train_sizes=train_sizes, scoring=mse_scorer)

    train_scores_mean = -np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = -np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Training MSE")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g", label="Cross-validation MSE")

    plt.legend(loc="best")
    return plt

title = "Learning Curves (Neural Network)"
plot_learning_curve_MSE(ANN_model, title, X_train, y_train, cv=5)
plt.show()

```
